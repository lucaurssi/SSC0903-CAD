```python
import matplotlib.pyplot as plt 
# pip install matplotlib
```

```python
# sequential
sy = [0.00013, 0.00004, 0.00002, 0.00024, 0.00023, 0.00024, 0.00023, 0.00026, 0.00025] 
# x-coordinates of left sides of bars  
left = [1, 2, 3, 4, 5, 6, 7, 8, 9] 
  
# labels for bars 
tick_label = ['seq', '2 P', '4 P', '6 P', '8 P', '10 P', '20 P', '32 P', '40 P']
  
# plotting a bar chart 
plt.bar(left, sy, tick_label = tick_label, 
        width = 0.8, color = ['red', 'green']) 
  
# naming the x-axis 
plt.xlabel('number of processors') 
# naming the y-axis 
plt.ylabel('Time') 
# plot title 
plt.title('MAT_SIZE = 100') 
  
# function to show the plot 
plt.show() 
```

```python
sy = [1, 0.00004, 0.00002, 0.00024, 0.00023, 0.00024] 
for i in range(len(sy)):
    sy[i] = 0.00013/sy[i]
x1 = [0,2,4,6,8,10] 

# plotting the line 1 points  
plt.plot(x1, sy, label = "") 
plt.plot([0,10], [0,10], label="",linestyle=":")
  

  
# naming the x axis 
plt.xlabel('number de processors') 
# naming the y axis 
plt.ylabel('Speedup') 
# plot title 
plt.title('MAT_SIZE = 100') 
  
  
# function to show the plot 
plt.show() 
```
